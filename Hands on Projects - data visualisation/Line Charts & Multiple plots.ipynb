```python
import pandas as pd 
```

```python
unrate = pd.read_csv('unrate.csv')
unrate.head()
```

```python
unrate['DATE'] = pd.to_datetime(unrate['DATE'])
```

```python
unrate.head(15)
```

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
plt.plot()
plt.show()
```

```python
x = unrate['DATE'].head(12)
y = unrate['VALUE'].head(12)
```

```python
plt.plot(x,y, color='green')
```

```python
plt.plot(x,y,color='orange')
plt.xticks(rotation = 90)
```

```python
plt.plot(x,y,color='violet')
plt.xticks(rotation=90)
plt.xlabel('Month')
plt.ylabel('Uemployment Rate')
plt.title('Monthly Unemplyment, 1948')
plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)
ax1.plot(unrate['DATE'].head(12),unrate['VALUE'].head(12))
ax2.plot(unrate.DATE.loc[12:24],unrate.VALUE.loc[12:24])
```

```python
fig = plt.figure(figsize=(14,12))
for i in range(4):
    ax = fig.add_subplot(1,5,i+1)
    start = i*12
    end = (i+1)*12
    subset = unrate[start:end]
    ax.plot(subset['DATE'], subset['VALUE'], color='maroon')
    plt.xticks(rotation=90)
```

```python
unrate['MONTH'] = unrate['DATE'].dt.month
```

```python
import seaborn as sns
fig = plt.figure(figsize=(6,6))

plt.plot(unrate[0:12]['MONTH'], unrate[0:12]['VALUE'], c='red', label='1948')
plt.plot(unrate[12:24]['MONTH'], unrate[12:24]['VALUE'], c='blue', label='1949')
plt.legend(loc='upper left')
sns.set_style('dark')
```

```python
fig = plt.figure(figsize=(12,6))
plt.plot(unrate[0:12]['MONTH'], unrate[0:12]['VALUE'], c='orange', label='1948')
plt.plot(unrate[12:24]['MONTH'], unrate[12:24]['VALUE'], c='red', label='1949')
plt.plot(unrate[24:36]['MONTH'], unrate[24:36]['VALUE'], c='blue', label='1950')
plt.plot(unrate[36:48]['MONTH'], unrate[36:48]['VALUE'], c='violet', label='1951')
plt.plot(unrate[48:60]['MONTH'], unrate[48:60]['VALUE'], c='green', label='1952')
plt.legend(loc='upper left')
plt.title('The unemplyment trends from year 1948 to 1952')
plt.xlabel('months')
plt.ylabel('unrate')
```
